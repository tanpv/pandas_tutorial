````python
import pandas as pd
df_imdb= pd.read_csv('./data/IMDB_Movie.csv')
df_imdb.head()
````

# . operator

```df.column_name```

Each column name could be consider as a dataframe attribute, so each column could be accessed with . operator like a object class 

````python
# select one column with . operator
df_imdb.director_name
````

# ```[]``` operator
```df[column selector]```
column selector could be;
    * a column
    * list of column use []
    * list of column use :
    

````python
# select multiple columns with [] operator
df_imdb[['director_name', 'movie_title']]
````

````python
# select all columns with [] operator
df_imdb[:].head()
````

# loc operator
```df.loc[[select row with row index],[select column with column index]]```
when print out a dataframe
* the top most row contain dataframe index for columns
* the left most column contain dataframe index for rows
* selector for row, column could be
    * one index
    * list of index ```[]```
    * use slicing ```:```

````python
# select firt row on fist column
df_imdb.loc[0,'color']
````

````python
# select firt 3 rows on 3 first columns use list
df_imdb.loc[[0,1,2],['color', 'director_name', 'num_critic_for_reviews']]
````

````python
# select first 3 rows on 3 first columns use slice
df_imdb.loc[:2, :'num_critic_for_reviews']
````

````python
# select all rows of all columns
df_imdb.loc[:,:].head()
````

# iloc operator

```df.iloc[[select columns with their position], [select rows with their position]]```

* imagine each row is assign a number, first row is assigned to 0 and final row is assigned to (total number of rows - 1)
* imagine each column is assign a number, first column is assign to 0 and final column is assigned to (total number of columns - 1)
* selector could be:
    * a number
    * a list of number
    * slicing with : operator
    

````python
# select first row on first column
df_imdb.iloc[0,0]
````

````python
# select first 3 rows on first 3 columns use list of number
df_imdb.iloc[[0,1,2],[0,1,2]]
````

````python
# select first 3 rows on first 3 columns use slicing
df_imdb.iloc[:3,:3]
````

````python
# select all rows of all columns
df_imdb.iloc[:,:].head()
````
