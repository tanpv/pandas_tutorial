# How to draw data in each day ?

# Resampling

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head(3)
```

```python
opsd_daily.shape
```

```python
opsd_weekly_mean.shape
```

```python
start, end = '2017-01', '2017-06'

fig, ax = plt.subplots()

ax.plot(opsd_daily.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       label='daily')

ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
        marker = '.',
        linestyle='-',
        label='weekly')

ax.legend()
```

```python
# resampling with monthly
opsd_monthly = opsd_daily[data_columns].resample('M').sum(min_count=28)
opsd_monthly.head(3)
```

```python
fig, ax = plt.subplots()
ax.plot(opsd_monthly['Consumption'], color='black', label='Monthly')
opsd_monthly[['Wind', 'Solar']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
```

```python
opsd_annual = opsd_daily[data_columns].resample('A').sum(min_count=360)
opsd_annual.tail()
```

```python
opsd_annual['Consumption'].plot()
```

# Rolling window

```python
opsd_7d = opsd_daily[data_columns].rolling(7, center=True).mean()
opsd_7d.tail(10)
```

```python
start, end = '2017-01', '2017-02'

# want to draw multiple plot at same time
fig, ax = plt.subplots()

ax.plot(opsd_daily.loc[start:end, 'Solar'],
        marker='.',
        linestyle='-',
        label='Daily')

# ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'],
#        marker='o',
#        linestyle='-',
#        label='Weekly mean')


ax.plot(opsd_7d.loc[start:end, 'Solar'],
       marker='.',
       linestyle='-',
       label='7d rolling mean')

ax.legend()
```

```python
# rolling to see the long term trend better ?
# draw multiple rolling will show how smoth will be when 
# doing rolling with long time
opsd_365d = opsd_daily[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
opsd_365d['Consumption'].plot()
```
