```python
import pandas as pd
df_imdb= pd.read_csv('./data/IMDB_Movie.csv')
df_imdb.head()
```

# Director with most social network facebook likes

```python
df_imdb.columns
df_fb_likes = df_imdb[['director_name', 'movie_title', 'movie_facebook_likes']]
df_fb_likes.groupby('director_name').sum().sort_values('movie_facebook_likes', ascending=False).head()
```
