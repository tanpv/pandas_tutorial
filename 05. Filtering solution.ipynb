```python
# import pandas and load data
import pandas as pd
df_imdb= pd.read_csv('./data/IMDB_Movie.csv')
df_imdb.head()
```

```python
# print out columns
df_imdb.columns
```

```python
# logical compare return a series of logical value
df_imdb['director_name'] == 'James Cameron'
```

```python
# filter with == logical
# find all james cameron movies
# select row which logical compare return True
df_imdb[df_imdb['director_name'] == 'James Cameron'][['director_name','movie_title']]
```

```python
# same way of working above, do other filter
# filter with one logical condition
# select movies with imdb_score > 9
df_imdb[df_imdb['imdb_score'] > 9]
```

```python
# filter with multiple logical condition
# select movies with duration > 90 and duration < 120
df_imdb[(df_imdb['duration'] < 120) & (df_imdb['duration'] > 90)]
```

```python
# search all action movie --> interesting question
# counting number of action movie
df_imdb.genres.head()
```

```python
def is_action(x):
    if 'Action' in x:
        return True
    else:
        return False

df_imdb.genres.apply(is_action)

df_imdb[df_imdb.genres.apply(is_action)][['movie_title', 'genres']]

# df_imdb.genres.apply(is_action).sum()
```
